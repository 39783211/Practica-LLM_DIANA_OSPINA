```python
# Seleccionar solo columnas útiles

columnas_utiles = [
    'Asunto',
    'Cuerpo',
    'De: (nombre)',
    'De: (dirección)',
    'Para: (nombre)',
    'Para: (dirección)'
]

df = df[columnas_utiles]

print(df.shape)
df.head()
```

```python
import re
import pandas as pd

def anonimizar_texto(texto):

    if pd.isna(texto):
        return texto

    texto = str(texto)

    # Correos electrónicos
    texto = re.sub(
        r'[\w\.-]+@[\w\.-]+',
        '[EMAIL]',
        texto
    )

    # NIT aproximado
    texto = re.sub(
        r'\b\d{8,12}\b',
        '[NIT]',
        texto
    )

    # Teléfonos
    texto = re.sub(
        r'\b\d{7,10}\b',
        '[TELEFONO]',
        texto
    )

    # Órdenes de trabajo OT
    texto = re.sub(
        r'OT[\s:-]*\d+',
        '[OT]',
        texto,
        flags=re.IGNORECASE
    )

    return texto
```

```python
columnas_anon = [
    'Asunto',
    'Cuerpo',
    'De: (nombre)',
    'Para: (nombre)'
]

for col in columnas_anon:
    df[col] = df[col].apply(anonimizar_texto)
```

```python
df["texto_caso"] = df["Asunto"].fillna("") + " " + df["Cuerpo"].fillna("")
df[["Asunto", "texto_caso"]].head()
```

```python
def clasificar_categoria(texto):
    texto = str(texto).lower()

    if "instal" in texto or "ot" in texto or "orden de trabajo" in texto:
        return "Implementación"
    elif "factura" in texto or "cobro" in texto or "nota crédito" in texto:
        return "Facturación"
    elif "caído" in texto or "falla" in texto or "indisponible" in texto:
        return "Soporte técnico"
    elif "traslado" in texto or "cambio" in texto:
        return "Traslado / Cambio"
    elif "portabilidad" in texto or "sim" in texto or "línea" in texto:
        return "Portabilidad móvil"
    else:
        return "Servicio al cliente"

df["categoria"] = df["texto_caso"].apply(clasificar_categoria)

df[["Asunto", "categoria"]].head()
```

```python
df["asunto_limpio"] = df["Asunto"].str.lower().str.replace("re:", "", regex=False).str.strip()

df["numero_escalamientos"] = df.groupby("asunto_limpio")["asunto_limpio"].transform("count")

df[["Asunto", "numero_escalamientos"]].head()
```

```python
def asignar_prioridad(row):
    if row["numero_escalamientos"] >= 4:
        return "Alta"
    elif row["numero_escalamientos"] >= 2:
        return "Media"
    else:
        return "Baja"

df["prioridad"] = df.apply(asignar_prioridad, axis=1)
```

```python
def asignar_semaforo(prioridad):
    if prioridad == "Alta":
        return "Rojo"
    elif prioridad == "Media":
        return "Amarillo"
    else:
        return "Verde"

df["semaforo"] = df["prioridad"].apply(asignar_semaforo)

df[["Asunto", "categoria", "numero_escalamientos", "prioridad", "semaforo"]].head()
```
